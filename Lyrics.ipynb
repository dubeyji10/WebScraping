# scraping lyrics of Far Caspian's A dream of You from Genius 

```python
import requests
from bs4 import BeautifulSoup as BS
```

```python
url = "https://genius.com/Far-caspian-a-dream-of-you-lyrics"
response = requests.get(url)
```

```python
html = response.content
```

```python
soup1 = BS(html,'html.parser')
```

```python
print(type(soup1))
```

```python
div1 = soup1.findAll('div')
```

```python
print(type(div1))
```

```python
print(div1)
```

```python
div2 = soup1.findAll("div", {"class": "lyrics"})
```

```python
type(div2)
```
